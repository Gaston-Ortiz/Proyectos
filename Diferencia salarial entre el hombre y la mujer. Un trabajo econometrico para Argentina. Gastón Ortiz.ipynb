<h2><center>Introducción</center></h2>
Uno de los temas más discutidos en el 2019 fue la diferencia salarial entre el hombre y la mujer.Si tenemos en cuenta que dos personas son exactamente iguales en cuanto a capacidades y en cuanto a la tarea que realizan, se piensa que deberían ganar lo mismo. O como sale en nuestra constitución en el artículo 14 “igual remuneración por igual tarea”.<br>
Teniendo en cuenta la encuesta permanente de hogares, en Argentina el salario promedio de un hombre es de \$19884  y el de una mujer es de  \$15172, pero ¿A qué se deben estas diferencias?<br>
Esta diferencia salarial puede deberse a que hombres y mujeres se diferencian en características individuales. El salario está influenciado por diferentes variables que lo afectan, como puede ser la experiencia y educación o por diferencias en las horas trabajadas. Aunque también, parte de esta brecha puede deberse a una discriminación salarial contra la mujer.<br>
En este trabajo se explica la diferencia salarial con el método de Oaxaca.

<h2><center>Metodología</center></h2>
Para tener una idea de a que nos referimos con discriminación salarial suponemos que un empleador puede elegir si contratar a un hombre o a una mujer donde ambos tiene las mismas cualidades y solo se diferencian por su sexo. Entonces, si el empleador no tiene nada en contra de ambos sexos, será indiferente a contratar un hombre o una mujer. En cambio sí al empleador le genera utilidad negativa contratar un mujer. Él, para compensar esto, estará dispuesto a pagarle menos que a un hombre, produciendo así una brecha salarial.<br> 

Para saber si estamos en presencia de una discriminación salarial contra la mujer tendremos que observar si el salario relativo del hombre excede al de la mujer para un mismo trabajo o de acuerdo a los mismos criterios de pago. Y a esta diferencia la dividimos por el salario relativo sin discriminación.
Para tener en cuenta esto, se define un coeficiente de discriminación de la siguiente manera: <br>
$D=(W_h/W_m -(W_h/W_m )^0) / (W_h/W_m )^0 $ (1)<br>
Donde $W_h$ es el salario observado del hombre y $W_m$ el salario observado de la mujer. Además  $W_h/W_m$  es el salario relativo observado y $(W_h/W_m )^0$  es el salario relativo en ausencia de discriminación.Si a la ecuación 1 le aplicamos el logaritmo natural y hacemos algunos arreglos matemáticos, esto sería igual a:<br> 
$ln⁡(D+1)=  ln⁡(W_h/W_m )-ln(W_h/W_m )^0$   (2)<br>
Además se asume que los empleadores en el mercador laboral sin discriminación cumplen con el principio de minimización de costos. Donde la el salario relativo en ausencia de discriminación es igual a productividad marginal del hombre dividida por la productividad marginal de la mujer.<br> 
$(W_h/W_m )^0=  MP_h/MP_m   $ (3)<br>
Donde $MP_h$ y $MP_m$ es igual a la productividad marginal de cada uno.<br>
Para poder estimar el coeficiente de discriminación de manera empírica, es necesario hacer algunos ajustes y supuestos. Como no conocemos el salario relativo sin discriminación para poder estimarlo, y con esto obtener D, suponemos que no existe discriminación por lo que podemos definir 2 opciones:<br>
<ul>
<li> Que la estructura salarial que enfrentan actualmente las mujeres, sería la referencia de no discriminación también se podría aplicar para hombres. Esto quiere decir que las mujeres estarían recibiendo, en promedio, el salario correspondiente a una situación sin discriminación y que los hombres estarían recibiendo más que el salario sin discriminación. La discriminación se presenta como la diferencia salarial entre el salario observado y salario que recibiría en una situación sin discriminación del hombre.</li>
<li> Que la estructura salarial que enfrentan los hombres sería la referencia de no discriminación y se podría aplicar también a mujeres. Esto significa que los hombres estarían recibiendo, en promedio, el salario sin discriminación y que las mujeres estarían recibiendo menos que este. Por lo tanto, la discriminación se presenta como la diferencia entre el salario sin discriminación y el salario observado de la mujer.</li>
</ul>
Para estimar la estructura salarial de cada uno se hará a través de una regresión por el método de mínimos cuadrados ordinarios, usando el programa Python. La ecuación del salario que se estimara para cada sexo será una función semi-logaritmica(los coeficiente si se los multiplica x 100 dan el cambio porcentual en la variable y) de la siguiente forma:<br>
$ln⁡(w_i )=βX_i^,+ u_i  $   i=1,2,…,n  (4)<br>
Donde $w_i$ es el salario de cada trabajador i, $X_i^,$, es el vector de características individuales, $β$ es el vector de coeficientes de cada característica y $u_i$ el término de error.<br>
Si tenemos en cuenta la ecuación (2) y los dos supuestos hechos podemos dividir la diferencia salarial en 2 términos, uno debido a las diferentes características y otro debido a la discriminación. Entonces si tenemos que  G  es la diferencia salarial, esta es igual a:<br>
$G=  (w ̅_h - w ̅_m)/w ̅_m$      (5)<br>
Donde $w ̅_h$  y $w ̅_m$ son el salario promedio para el hombre y la mujer respectivamente. Si a esto le aplicamos el logaritmo natural y hacemos algunos arreglos tenemos que:<br>
$ln⁡(G-1)=ln⁡(w ̅_h )-ln⁡(w ̅_m )$   (6)<br>
Y por propiedades de MCO los términos de la derecha son igual a:<br>
$ln⁡(w ̅_h )=\hat{β} ̂_h  \bar{X}_h^, $    (7)<br>
$ln⁡(w ̅_m )=\hat{β} ̂_m  \bar{X}_m^, $  (8)<br>
Donde $\bar{X}_h^,$ y $\bar{X}_m^,$son los valores medios de cada uno de los regresores de hombre y mujer respectivamente. Además $\hat{β} ̂_h$ y $\hat{β} ̂_m$ son los coeficientes estimados de cada uno de los regresores. Con esto obtenemos la estructura de determinación del salario de cada uno.<br>
Por sustitución tenemos que:<br>
$ln⁡(G-1)= \hat{β}_h  \bar{X}_h^,- \hat{β}_m  \bar{X}_m^, $     (9)<br>
Si tenemos que en cuenta que:<br>
$∆\bar{X}^, ̅=  \bar{X}_h^,- \bar{X}_m^,$   (10)<br>
$∆\hat{β}= \hat{β}_m  - \hat{β}_h$    (11)<br>
Sustituyendo $\hat{β}_h= \hat{β}_m  - ∆\hat{β}$ en (8) y  haciendo uno arreglos matemáticos, nos quedaría:<br>
$ln⁡(G-1)= ∆\bar{X}^,\hat{β}_m - \bar{X}_h^,∆\hat{β}$  (12)<br>
Si asumimos que la actual estructura de determinación de salario de la mujer es la que se podría aplicar a un mercado laboral sin discriminación, podemos demostrar que:<br>
$ln(\hat{⁡w_h}/\hat{w_m})^0= ∆\bar{X}^,\hat{β}_m $  (13)<br>
$ln⁡(\hat{D}+1)= -\bar{X}_h^, ̅∆\hat{β} $(14)<br>
La primera expresión representa la descomposición de la diferencia salarial que se debe a diferencias en las características individuales. La segunda expresión se refiere a la parte de los efectos discriminatorios. Con esto tenemos todo lo necesario para calcular $\hat{D}$.<br>
Si suponemos que la estructura salarial del hombre es la predomina en un mercado laboral sin discriminación, deberíamos sustituir el término $\hat{β}_m= \hat{β}_h+ ∆\hat{β}$ en vez de $\hat{β}_h= \hat{β}_m - ∆\hat{β}$ por lo que nos quedaría lo siguiente:<br>
$ln⁡(G-1)= ∆\bar{X}^,\hat{β}_h - \bar{X}_m^,∆\hat{β}$ (15)<br>
Por lo que podemos demostrar que:<br>
$ln⁡(\hat{w_h}/\hat{w_m})^0 = ∆\bar{X}^,\hat{β}_h$  (16)<br>
$ln⁡(\hat{D}+1)= -\bar{X}_m^,∆\hat{β}$  (17)<br>
Al igual que lo dicho antes la primera expresión representa la descomposición de la diferencia salarial que se debe a diferencias en las características individuales. La segunda expresión se refiere a la parte de los efectos discriminatorios. Con esto también podemos calcular $\hat{D}$.
Para calcular la estructura de determinación del salario para los dos sexos respectivamente como base de datos utilizaremos la encuesta permanente de hogares (eph) del INDEC del cuarto trimestre del 2018.<br>
A partir de la eph crearemos nuestras variables para poder estimar por medio del método mínimos cuadros ordinarios la estructura de la determinación del salario. La variable dependiente es la siguiente:<br>
<ul>
<li><strong>Logaritmo natural del salario</strong>(LWAGE): Nuestra variable endógena, es igual al salario total de la ocupación principal del encuestado y le aplicamos logaritmo natural.</li>
</ul>
Variables independientes:
<ul>
    <li> <strong>Experiencia</strong> (EXPER): Experiencia laboral de la persona encuestada. se ha creado una variable proxy con la siguiente formula:
$EXPER = AÑOS – EDUCACION – 6$ (18).Además se agrega la variable experiencia al cuadrado (EXPER2).</li>

En cambio, sí creemos que la discriminación se da en algunos tipos de industrias y ocupaciones, agregar estas variables al modelo será lo recomendable. Al obtener los coeficientes estimados para hombres y mujeres. Obtuvimos el coeficiente de discriminación tomando como estructura de determinación del salario sin discriminación a los coeficientes estimados de la regresión de la mujer, a través de la siguiente formula:<br> 
$ln⁡(G-1)-∆\bar{X}\hat{β}_m = ln⁡(D+1)$ (19)<br>
Donde $∆\bar{X} = \bar{X}_h - \bar{X}_m$ y $ln⁡(D+1)= -\bar{X}_h∆\hat{β}$.<br>
Por otra parte, si tomamos como estructura para determinar el salario sin discriminación a los coeficientes estimados de la regresión del hombre, sería con la siguiente formula:<br>
$ln⁡(G-1) - ∆\bar{X}\hat{β}_h = ln⁡(D+1)$ (20)<br>
Donde $∆\bar{X} = \bar{X}_h - \bar{X}_m$ y $ln⁡(D+1)= -\bar{X}_m∆\hat{β}$.<br>
Usando este método podemos calcular el efecto de la discriminación como el residual que queda después de ajustar por diferencia en varias características la diferencia salarial entre sexos.<br>

```python
# Obtengo la diferencia salarial 
LNG1 = dfh.LWAGE.mean() -  dfm.LWAGE.mean() 
LNG1
```

Definimos una función que reciba como parametros 2 data frame, uno para hombres y otro para mujeres. Ademas recibe un tercer parametro que son los coeficientes betas estimados de la determinación del salario.<br>
Hay que tener en cuenta que los betas tienen que ser los que supongamos que prevalecen en ausencia de discriminación<br>
Lo que devuelve esta función es el logaritmo natural de D + 1 y un data frame con dos columnas, la primera  con los ajustes de las diferencias salariales en cada variables y la segunda con los porcentajes de que reprensentan sobre la diferencia salarial.

```python
def coeficientes(dataH,dataM,betas):
    coef = pd.DataFrame(index=[np.insert(betas.index.values,0,"Diferencia Salarial")],columns=["Discriminación","Porcentaje"])
    LN_G1 = dataH.LWAGE.mean() -  dataM.LWAGE.mean() 
    coef.loc["Diferencia Salarial","Discriminación"] = LN_G1
    for i in betas:
        var = betas[betas == i].index[0]
        coef.loc[var,"Discriminación"] = (dataH[var].mean() - dataM[var].mean()) * betas[var]
    suma = 0
    for i in coef["Discriminación"][1:]:
        suma += i
    LN_D1 = LN_G1 - suma
    coef.loc["LN_D1","Discriminación"] = LN_D1
    for i in coef["Discriminación"]:
        coef.loc[coef["Discriminación"] == i,"Porcentaje"] = i / LN_G1 * 100
    coef.loc["D","Discriminación"] = np.exp(LN_D1) - 1
    return coef, LN_D1
```

```python
# Obtengo el LN(D+1) si suponemos que en ausencia de discrimacion prevalece la estructura de determinacion del salario de la mujer 
ajustes_m, ln_D_m = coeficientes(dfh,dfm,mm.params.iloc[1:])
print("LN(D+1) si suponemos que la estructura de determinacion del salario de la mujer es la que prevalece",round(ln_D_m,3))
# Obtengo el LN(D+1) si suponemos que en ausencia de discrimacion prevalece la estructura de determinacion del salario del hombre 
ajustes_h, ln_D_h = coeficientes(dfh,dfm,mh.params.iloc[1:])
print("LN(D+1) si suponemos que la estructura de determinacion del salario del hombre es la que prevalece",round(ln_D_h,3))
```

```python
def cuadros(ajustes):
    cuadro_ampliado = ajustes.iloc[:12].copy()
    cuadro_ampliado.loc["Industria","Discriminación"] = ajustes.iloc[12:25].sum()[0]
    cuadro_ampliado.loc["Industria","Porcentaje"] = ajustes.iloc[12:25].sum()[1]
    cuadro_ampliado.loc["Ocupación","Discriminación"] = ajustes.iloc[25:-2].sum()[0]
    cuadro_ampliado.loc["Ocupación","Porcentaje"] = ajustes.iloc[25:-2].sum()[1]
    cuadro_ampliado = pd.concat([cuadro_ampliado,ajustes.iloc[-2:]])
    return cuadro_ampliado
```

<h2><center>Conclusiones</center></h2>
Analizando ambas regresiones que tienen en cuenta las características personales y también el tipo de ocupación e industria, se puede sugerir que puede existir un cierto grado de discriminación salarial contra la mujer, por lo que no se estaría respetando el artículo 14 de nuestra constitución que dice “igual remuneración por igual tarea”.<br>
También hay que tener en cuenta que este método cuenta con algunas desventajas como puede ser que no podemos medir la cantidad de hijos de cada encuestado, esto influye en que una madre al tener un hijo sale del mercado laboral, por lo que no estaría ganando experiencia, además si suponemos que pasa demasiado tiempo fuera comenzara a depreciarse la experiencia ganada anteriormente, por lo que hay un doble efecto de salir de la fuerza laboral que no puede ser medido. Si esto se pudiera medir correctamente la variable experiencia disminuiría el coeficiente de discriminación mucho más 
Otro factor a tener en cuenta es si la estructura salarial del hombre y la mujer son iguales, esto afectaría a como se determina el salario de cada uno. Ya que hombres y mujeres son diferentes por lo que no podemos pretender que su estructura para determinar el salario sea igual.<br>
A pesar de estas desventajas en nuestro modelo, hay un gran porcentaje de la diferencia salarial que no puedo ser explicado por diferencias de características entre hombre y mujer. Esto puede deberse en parte a la discriminación salarial que sufre la mujer y también en parte a factores no observados que no podemos medir. Por lo que se debería seguir investigando en esta temática  para poder llegar a una solución al problema de discriminación salarial contra la mujer que hemos encontrado.<br>
